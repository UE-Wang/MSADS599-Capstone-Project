```python
# Convert 'season' column to numeric
nfl['season'] = nfl['season'].astype(int)
```

```python
# Separate regular weeks and playoff rounds
regular_weeks = nfl[nfl['week'].str.isdigit()]
playoff_rounds = nfl[~nfl['week'].str.isdigit()]
```

```python
# Group by 'season' and count rows
season_counts = nfl.groupby('season').size()

print(season_counts)
```

```python
print("Number of columns in the DataFrame:", nfl.shape[1])

print("\nList of columns:\n")
print(nfl.columns.tolist())
```

```python
# count of missing values

nfl.isna().sum()
```

```python
# Create a new feature 'home_win'
nfl['home_win'] = np.where(nfl['score_home'] > nfl['score_away'], 1, 0)
```

```python
# Calculate third down conversion rate for the home and away team
nfl['third_down_rate_home'] = nfl['third_down_comp_home']/nfl['third_down_att_home']*100
nfl['third_down_rate_away'] = nfl['third_down_comp_away']/nfl['third_down_att_away']*100

# Calculate fourth down conversion rate for the home and away team
nfl['fourth_down_rate_home'] = nfl['fourth_down_comp_home'] / nfl['fourth_down_att_home'] * 100
nfl['fourth_down_rate_away'] = nfl['fourth_down_comp_away'] / nfl['fourth_down_att_away'] * 100
```

```python
# Calculate pass completion rate for the home and away team
nfl['pass_rate_home'] = nfl['pass_comp_home'] / nfl['pass_att_home'] * 100
nfl['pass_rate_away'] = nfl['pass_comp_away'] / nfl['pass_att_away'] * 100
```

```python
# Calculate redzone conversion rate for the home and away team
nfl['redzone_rate_home'] = nfl['redzone_comp_home'] / nfl['redzone_att_home'] * 100
nfl['redzone_rate_home'] = nfl['redzone_comp_away'] / nfl['redzone_att_away'] * 100
```

```python
# Convert time string to seconds
def time_to_seconds(time_str):
    minutes, seconds = map(int, time_str.split(':'))
    return minutes * 60 + seconds

# Apply the function to convert time strings to seconds
nfl['possession_away_seconds'] = nfl['possession_away'].apply(time_to_seconds)
nfl['possession_home_seconds'] = nfl['possession_home'].apply(time_to_seconds)
```

# Exploratory Data Analysis

```python

# Create a list of numerical columns
numeric_columns = nfl.select_dtypes(include=['int64']).columns

fig, axs = plt.subplots(5, 10, figsize=(20, 20))

axs = axs.flatten()

# Plot each numerical column
for i, column in enumerate(numeric_columns):
    ax = axs[i] 
    ax.boxplot(nfl[column], vert=True) 
    ax.set_title(column, fontsize=10) 
    ax.set_xticks([]) 
    ax.set_ylabel(column) 

plt.tight_layout()  
plt.show()
```

```python
home_win_counts = nfl['home_win'].value_counts()

# Plotting the bar plot
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='home_win', data=nfl, palette=['orange', 'skyblue'])

# Calculate the percentage for each category
total_count = nfl.shape[0]
for i in ax.patches:
    percentage = '{:.1f}%'.format(100 * i.get_height() / total_count)
    x = i.get_x() + i.get_width() / 2
    y = i.get_height() + 10
    ax.annotate(percentage, (x, y), ha='center')

plt.title('Count of Home Wins')
plt.xlabel('Home Win (0: No, 1: Yes)')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot for Yards Home vs Score Home
sns.lineplot(data=nfl, x='yards_home', y='score_home', hue='home_win', palette='coolwarm', marker='o', ci=None, ax=axes[0])
axes[0].set_title('Yards Home vs Score Home')
axes[0].set_xlabel('Yards Home')
axes[0].set_ylabel('Score Home')
axes[0].legend(title='Home Win', loc='upper left')
axes[0].grid(True)

# Plot for First Downs Home vs Score Home
sns.lineplot(data=nfl, x='first_downs_home', y='score_home', hue='home_win', palette='coolwarm', marker='o', ci=None, ax=axes[1])
axes[1].set_title('First Downs Home vs Score Home')
axes[1].set_xlabel('First Downs Home')
axes[1].set_ylabel('Score Home')
axes[1].legend(title='Home Win', loc='upper left')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
#Completion Efficiency Rates

def plot_line(ax, data, x_col, y_col, hue_col, title):
    sns.lineplot(data=data, x=x_col, y=y_col, hue=hue_col, palette='coolwarm', ax=ax)
    ax.set_title(title)
    ax.set_xlabel(x_col)
    ax.set_ylabel(y_col)
    ax.legend(title=hue_col.capitalize() + ' Win', loc='upper left')
    ax.grid(True)

# Create 2 * 2 subplots
fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# Line plot for third down rate
plot_line(axes[0, 0], nfl, 'third_down_rate_home', 'score_home', 'home_win', 'Third Down Rate Home vs Score Home')

# Line plot for fourth down rate
plot_line(axes[0, 1], nfl, 'fourth_down_rate_home', 'score_home', 'home_win', 'Fourth Down Rate Home vs Score Home')

# Line plot for pass rate
plot_line(axes[1, 0], nfl, 'pass_rate_home', 'score_home', 'home_win', 'Pass Rate Home vs Score Home')

# Line plot for redzone rate
plot_line(axes[1, 1], nfl, 'redzone_rate_home', 'score_home', 'home_win', 'Redzone Rate Home vs Score Home')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 6))

# Plot 1: Fumbles Home vs Average Score Home
plt.subplot(1, 2, 1)  
sns.lineplot(data=nfl, x='fumbles_home', y='score_home', estimator='mean')
plt.title('Average Score Home vs Fumbles Home')
plt.xlabel('Fumbles Home')
plt.ylabel('Average Score Home')
plt.grid(True)

# Plot 2: Interceptions Home vs Average Score Home
plt.subplot(1, 2, 2)
sns.lineplot(data=nfl, x='interceptions_home', y='score_home', estimator='mean')
plt.title('Average Score Home vs Interceptions Home')
plt.xlabel('Interceptions Home')
plt.ylabel('Average Score Home')
plt.grid(True)

plt.tight_layout() 
plt.show()
```

```python
# Group by 'season' and calculate mean for 'score_away' and 'score_home'

seasonal_data = nfl.groupby('season')[['score_away', 'score_home']].mean()

plt.figure(figsize=(12, 6))

sns.lineplot(data=seasonal_data['score_away'], color='blue', marker='o', label='Score Away')
sns.lineplot(data=seasonal_data['score_home'], color='red', marker='^', label='Score Home')

plt.title('Seasonal Trends: Average Score Away vs Score Home')
plt.xlabel('Season')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Weekly trends
def plot_seasonal_trends(data, title):
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=data['score_away'], color='blue', marker='o', label='Score Away')
    sns.lineplot(data=data['score_home'], color='red', marker='^', label='Score Home')
    plt.title(title)
    plt.xlabel('Season')
    plt.ylabel('Average Score')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot seasonal trends for regular weeks
gr_regular_weeks = regular_weeks.groupby('season')[['score_away', 'score_home']].mean()
plot_seasonal_trends(gr_regular_weeks, 'Regular Weeks: Seasonal Trends - Average Score Away vs Score Home')

# Plot seasonal trends for playoff rounds
gr_playoff_rounds = playoff_rounds.groupby('season')[['score_away', 'score_home']].mean()
plot_seasonal_trends(gr_playoff_rounds, 'Playoff Rounds: Seasonal Trends - Average Score Away vs Score Home')
```

```python
# Calculate mean Time of Possession for home win and home loss

mean_possession_home_win = nfl[nfl['home_win'] == 1]['possession_home_seconds'].mean()
mean_possession_home_lost = nfl[nfl['home_win'] == 0]['possession_home_seconds'].mean()

plt.figure(figsize=(10, 6))

bar_width = 0.4
bar_positions = [0, 1]

plt.bar(bar_positions, [mean_possession_home_win, mean_possession_home_lost], 
        width=bar_width, color=['skyblue', 'salmon'])

plt.text(bar_positions[0], mean_possession_home_win + 20, 
         f'{mean_possession_home_win:.2f}', ha='center')
plt.text(bar_positions[1], mean_possession_home_lost + 20, 
         f'{mean_possession_home_lost:.2f}', ha='center')

plt.ylabel('Mean Time of Possession (seconds)')
plt.title('Mean Time of Possession vs. Game Outcome (for Home Team)')
plt.xticks(bar_positions, ['Home Win', 'Home Loss'])
plt.grid(axis='y')

plt.show()
```

```python
# Calculate mean drives at home for home win and home loss

mean_drives_home_win = nfl[nfl['home_win'] == 1]['drives_home'].mean()
mean_drives_home_loss = nfl[nfl['home_win'] == 0]['drives_home'].mean()

plt.figure(figsize=(10, 6))

bar_width = 0.4
bar_positions = [0, 1]

bars = plt.bar(bar_positions, [mean_drives_home_win, mean_drives_home_loss], 
        width=bar_width, color=['skyblue', 'salmon'])

for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, 
             f'{bar.get_height():.2f}', ha='center')

plt.ylabel('Mean Number of Drives at Home')
plt.title('Mean Number of Drives at Home vs. Game Outcome (for Home Team)')
plt.xticks(bar_positions, ['Home Win', 'Home Loss'])
plt.grid(axis='y')

plt.show()
```

```python

# Sort the teams based on the number of wins
team_performance_home = nfl.groupby('home')['home_win'].sum().reset_index()
team_performance_home = team_performance_home.sort_values(by='home_win', ascending=False)
team_performance_away = nfl.groupby('away')['home_win'].apply(lambda x: (x == 0).sum()).reset_index()
merged_performance = pd.merge(team_performance_home, team_performance_away, how='outer', left_on='home', right_on='away')
merged_performance.drop('away', axis=1, inplace=True)
merged_performance.rename(columns={'home': 'team', 'home_win_x': 'home_wins', 'home_win_y': 'away_wins'}, inplace=True)
merged_performance['total_wins'] = merged_performance['home_wins'] + merged_performance['away_wins']

print(merged_performance)
```

```python

# Sort the nfl based on total wins
merged_performance_sorted = merged_performance.sort_values(by='total_wins', ascending=False)

plt.figure(figsize=(12, 6))

x = np.arange(len(merged_performance_sorted))

width = 0.35

# Create home wins bars
bars_home = plt.bar(x - width/2, merged_performance_sorted['home_wins'], width, label='Home Wins', color='skyblue')

# Create away wins bars, stacked on top of home wins bars
bars_away = plt.bar(x + width/2, merged_performance_sorted['away_wins'], width, label='Away Wins', color='lightgreen')

plt.title('Home and Away Wins by Team')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(x, merged_performance_sorted['team'], rotation=45, ha='right')
plt.legend()

for bar in bars_home + bars_away:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, height, ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
